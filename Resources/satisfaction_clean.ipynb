```python
import pandas as pd
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('airline_passenger_satisfaction.csv')

df.head()
```

```python
df.info()
```

```python
# Drop the `Unnamed: 0` column
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Drop null values in `arrival_delay_in_minutes`
df.dropna(subset = ['arrival_delay_in_minutes'], inplace=True)
```

```python
# Convert `satisfaction` column to binary
df['satisfaction'] = df['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

```python
# Print the count of occurances of values in the `satisfaction` column
print("\nCount of values in satisfaction column: \n", df['satisfaction'].value_counts().to_markdown(numalign="left", stralign="left"))
```
