```python
import pandas as pd
```

```python
df = pd.read_csv('us-dealers-used.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pd.set_option('display.max_rows', None)

make_counts = df['make'].value_counts()
make_counts
```

```python
df.dropna(inplace=True)
```

```python
# Drop specified columns
columns_to_drop = ['id', 'vin', 'stock_no', 'seller_name','street','zip']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.shape[0]
```

```python
df_filtered = df[df['year'] >= 2017]
```

```python
makes_to_drop = make_counts[make_counts < 10489].index
```

```python
df_filtered = df_filtered[~df_filtered['make'].isin(makes_to_drop)]
```

```python
df_sampled = df_filtered.sample(n=150000, random_state=42)
```

```python
df_sampled.head(25)
```

```python
df_sampled.to_csv('used_clean.csv', index=False)
```
