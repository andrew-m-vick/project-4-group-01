```python
import pandas as pd
```

```python
df = pd.read_csv('us-dealers-used.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pd.set_option('display.max_rows', None)

make_counts = df['make'].value_counts()
make_counts
```

```python
df_tesla = df[df['make'].str.lower() == 'tesla'].copy()
```

```python
df_tesla['transmission'] = df_tesla['transmission'].fillna('Automatic')
df_tesla['fuel_type'] = df_tesla['fuel_type'].fillna('Electric')
df_tesla['engine_block'] = df_tesla['engine_block'].fillna('E')
df_tesla['engine_size'] = df_tesla['engine_size'].fillna('E')
```

```python
df_tesla.dropna(inplace=True)
```

```python
# Convert `miles` and `year` columns to integer
df_tesla['miles'] = df_tesla['miles'].astype(int)
df_tesla['year'] = df_tesla['year'].astype(int)
```

```python
# Drop specified columns
columns_to_drop = ['id', 'vin', 'stock_no', 'seller_name','street']
df_tesla.drop(columns=columns_to_drop, inplace=True)
```

```python
df_tesla_filtered = df_tesla[df_tesla['year'] >= 2017]
```

```python
df_tesla_filtered = df_tesla_filtered[df_tesla_filtered['miles'] >= 500]
```

```python
# Drop duplicate rows based on the 'vin' column, keeping the first occurrence
df_cleaned = df.drop_duplicates(subset=['vin'], keep='first').copy() 
```

```python
df_cleaned.dropna(inplace=True)
```

```python
# Convert `miles` and `year` columns to integer
df_cleaned['miles'] = df_cleaned['miles'].astype(int)
df_cleaned['year'] = df_cleaned['year'].astype(int)
```

```python
# Drop specified columns
columns_to_drop = ['id', 'vin', 'stock_no', 'seller_name','street']
df_cleaned.drop(columns=columns_to_drop, inplace=True)
```

```python
df_cleaned.shape[0]
```

```python
df_filtered = df_cleaned[df_cleaned['year'] >= 2017]
```

```python
# Filter the DataFrame to keep only rows where `miles` is less than or equal to 300000
df_filtered = df_filtered[df_filtered['miles'] <= 300000]
```

```python
df_filtered = df_filtered[df_filtered['miles'] >= 500]
```

```python
df_filtered = df_filtered[df_filtered['price'] >= 7000]
```

```python
# List of all US states
us_states = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
```

```python
# Filter out non-US states
df_filtered = df_filtered[df_filtered['state'].isin(us_states)]
```

```python
makes_to_drop = make_counts[make_counts < 10489].index
```

```python
df_filtered = df_filtered[~df_filtered['make'].isin(makes_to_drop)]
```

```python
df_filtered = pd.concat([df_filtered, df_tesla_filtered], ignore_index=True)
```

```python
df_sampled = df_filtered.sample(n=150000, random_state=42)
```

```python
df_sampled.to_csv('used_clean.csv', index=False)
```
